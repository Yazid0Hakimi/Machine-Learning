```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as skl
```

```python
file_path = r'dataSets\titanic.csv'
data = pd.read_csv(file_path)
data
```

```python
data.drop(columns=['Name'],inplace=True)
```

```python
data["Sex"].replace({"male":0 , "female":1}, inplace=True)
data
```

```python
data.info()
```

```python
survived = data[data['Survived'] == 1]
unsurvived = data[data['Survived'] == 0]
survived
```

```python
unsurvived
```

```python
sns.pairplot(data, hue='Survived')
plt.show()
```

```python
sns.scatterplot(x='Survived', y='Age',hue='Sex', data=data)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Sex', hue='Survived', data=data, palette='Set1')
plt.title('Survivors by Sex')
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Pclass', hue='Survived', data=data, palette='Set1')
plt.title('Survivors by Sex')
plt.xlabel('Class')
plt.ylabel('Count')
```
